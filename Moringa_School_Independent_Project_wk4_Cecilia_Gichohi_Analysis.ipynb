```
# Making the dataframe from csv file 
import pandas as pd
outliers = pd.read_csv("cleaned_data") 

day= pd.DataFrame(outliers)

outliers.head(20)
```

```
#using the IQR method
Q1=day.quantile(0.25)
Q2=day.quantile(0.50)
Q3=day.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```
RANGE1=Q1-(1.5*IQR)
RANGE2=Q3+(1.5*IQR)
RANGE1
RANGE2
```
